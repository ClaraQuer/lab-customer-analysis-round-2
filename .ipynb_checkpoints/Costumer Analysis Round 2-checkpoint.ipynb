```python
import pandas as pd 
import numpy as np
```

```python
mca = pd.read_csv('files_for_lab/csv_files/marketing_customer_analysis.csv')
mca.shape
```

```python
mca.columns
```

```python
mca = mca.drop(['Unnamed: 0'], axis = 1)
```

```python
mca.columns
```

```python
mca = mca.rename(columns={'EmploymentStatus': 'Employment Status',
                         'Number of Open Complaints': 'Number Of Open Complaints',
                         'Number of Policies': 'Number Of Policies'})


mca.head()
```

```python
mca = mca[['Customer', 'State', 'Gender', 'Marital Status','Education', 'Employment Status', 'Income', 
     'Customer Lifetime Value', 'Response', 'Coverage','Effective To Date','Location Code',  'Monthly Premium Auto',
     'Months Since Last Claim', 'Months Since Policy Inception',
     'Number Of Open Complaints', 'Number Of Policies', 'Policy Type',
     'Policy', 'Renew Offer Type', 'Sales Channel', 'Total Claim Amount',
     'Vehicle Class', 'Vehicle Size', 'Vehicle Type']]

mca.head()
```

```python
mca.dtypes
```

```python
mca._get_numeric_data()
```

```python
mca.select_dtypes(['object'])
```

```python
mca.duplicated().sum()
mca = mca.drop_duplicates()
mca.isna().sum()
```

```python
pd.set_option('display.max_columns', None)

mca[mca['State'].isna() & mca['Response'].isna()] #The ones with no response have no State Assigned. Maybe that is because they coulddn't ask.

mca[(mca['Months Since Last Claim'].isna() & mca['Number Of Open Complaints'].isna())] #The ones with no data about open complaints also have no data for previous complaints, suggesting they are new costumers.
```

```python
mca['Months Since Last Claim'] = mca['Months Since Last Claim'].fillna(0) #We change the values to 0 because if there is no recolection of a previous complaint, it is the same as if you had a 0.
mca['Number Of Open Complaints'] = mca['Number Of Open Complaints'].fillna(0) #We do the same here for the same reason.
mca['Response'] = mca['Response'].fillna('No') #If you don't know if someone responded, they probably didn't, so we change it to "No".
mca.isna().sum()
```

```python
mca['State'].value_counts()/(len(mca['State']))*100 #We check if there is a vast majority in one state big enough to assume the NAN are from there aswell. We see this is not the case.
```

```python
mca['State'] = mca['State'].fillna('Unknown') #We decide to change them to 'Unkown' because we can't know for sure.
mca.isna().sum()
```

```python
mca['Vehicle Class'].value_counts()/(len(mca['Vehicle Class']))*100 #Since there are so many different types, we will assign "Unkown" to the NaNs.
```

```python
mca['Vehicle Size'].value_counts()/(len(mca['Vehicle Size']))*100 #We will assign "Unkown" to the NaNs for the same reason as before. We think assuming they are Medium just because the majority are is not right.
```

```python
mca['Vehicle Type'].value_counts()/(len(mca['Vehicle Type']))*100 #Since there is only one type, and it is "A", we are assuming this means "Automatic car", and since the data is from the US, we will just assume they are all automatic.
```

```python
mca['Vehicle Class'] = mca['Vehicle Class'].fillna('Unknown')
mca['Vehicle Size'] = mca['Vehicle Size'].fillna('Unknown')
mca['Vehicle Type'] = mca['Vehicle Type'].fillna('A')

mca.isna().sum() #As we can see, there are no more NaN values.
```
